```python
# TASK 3: Calculate the distances between a launch site to its proximities
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
# Create a marker with distance to a closest city, railway, highway, etc.
# Draw a line between the marker to the launch site
# 1. Define Coordinates for nearby points (Example for CCAFS SLC-40)
launch_site = [28.563197, -80.576820]
railway = [28.57205, -80.58527]
highway = [28.56321, -80.57079]
city = [28.10473, -80.64531] # Melbourne, FL

# List of targets to loop through
targets = [
    {'name': 'Railway', 'coord': railway},
    {'name': 'Highway', 'coord': highway},
    {'name': 'City', 'coord': city}
]

for target in targets:
    # 2. Calculate Distance
    dist = calculate_distance(launch_site[0], launch_site[1], target['coord'][0], target['coord'][1])
    
    # 3. Add Distance Marker
    distance_marker = folium.Marker(
        target['coord'],
        icon=DivIcon(
            icon_size=(20,20),
            icon_anchor=(0,0),
            html='<div style="font-size: 12; color:#252526;"><b>%s</b></div>' % "{:10.2f} KM".format(dist),
        )
    )
    site_map.add_child(distance_marker)
    
    # 4. Draw PolyLine
    line = folium.PolyLine(locations=[launch_site, target['coord']], weight=1, color='green')
    site_map.add_child(line)

# Display Map
site_map
```

1. Are launch sites in close proximity to railways?
Finding: Yes (Typically < 2 km).
Explanation: Heavy rocket components, such as the Falcon 9 first stage and large fuel tanks, are too massive for standard road transport over long distances. Proximity to railways allows for the efficient delivery of these heavy materials directly from the manufacturing facilities.
2. Are launch sites in close proximity to highways?
Finding: Yes (Typically < 1.5 km).
Explanation: Easy access to highways is essential for the daily movement of personnel, smaller equipment, and emergency services. It ensures the site is logistically connected to the surrounding infrastructure.
3. Are launch sites in close proximity to the coastline?
Finding: Yes (Typically < 1 km).
Explanation: This is the most critical safety feature. Launching over the ocean ensures that if a mission fails, debris falls into the water rather than onto populated areas. It also provides a clear path for boosters to land on drone ships stationed at sea.
4. Do launch sites keep a certain distance away from cities?
Finding: Yes (Typically > 20 km).
Explanation: Launch sites are intentionally kept far from major cities to protect the public from the extreme noise levels (acoustic vibration) and the potential risk of explosions or toxic fuel leaks during a launch anomaly.
